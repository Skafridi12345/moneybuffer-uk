# Generate Synthetic Households

This notebook generates fictional MoneyBuffer UK household and transaction data for local exploration. The data is synthetic and must not be treated as real customer data.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.insert(0, str(PROJECT_ROOT / "src"))

from moneybuffer.data_generation import generate_households, generate_transactions
```

```python
households = generate_households(n=100, seed=42)
transactions = generate_transactions(households, months=6, seed=42)

households.head()
```

```python
transactions.head()
```

```python
output_dir = PROJECT_ROOT / "data" / "synthetic"
output_dir.mkdir(parents=True, exist_ok=True)

households.to_csv(output_dir / "households.csv", index=False)
transactions.to_csv(output_dir / "transactions.csv", index=False)

output_dir
```
